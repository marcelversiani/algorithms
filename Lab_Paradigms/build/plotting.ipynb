### Importing Libraries and Files

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
column_names = ["value", "time_in_ms", "coins_number","number_of_recursive_calls"]

troco_DC40 = pd.read_csv("../build/troco_DC40.csv", header=None, names=column_names)
troco_DC40_last = troco_DC40.iloc[40]
troco_DC40 = troco_DC40[:-1]

troco_GR_BR200 = pd.read_csv("../build/troco_GR_BR200.csv", header=None, names=column_names)
troco_GR_BR200_last = troco_GR_BR200.iloc[200]
troco_GR_BR200 = troco_GR_BR200[:-1]

troco_GR_GRLoses = pd.read_csv("../build/troco_GR_GRLoses.csv", header=None, names=column_names)
troco_GR_GRLoses_last = troco_GR_GRLoses.iloc[200]
troco_GR_GRLoses = troco_GR_GRLoses[:-1]

troco_PD_BR200 = pd.read_csv("../build/troco_PD_BR200.csv", header=None, names=column_names)
troco_PD_BR200_last = troco_PD_BR200.iloc[200]
troco_PD_BR200 = troco_PD_BR200[:-1]

troco_PD_GRLoses = pd.read_csv("../build/troco_PD_GRLoses.csv", header=None, names=column_names)
troco_PD_GRLoses_last = troco_PD_GRLoses.iloc[200]
troco_PD_GRLoses = troco_PD_GRLoses[:-1]
```

### Total Execution Time for Each Algorithm

```python
algorithms = ["DC40", "GR200", "PD200", "GR_GRLoses", "PD_GRLoses"]
execution_times = [troco_DC40_last["time_in_ms"], troco_GR_BR200_last["time_in_ms"], troco_PD_BR200_last["time_in_ms"], troco_GR_GRLoses_last["time_in_ms"], troco_PD_GRLoses_last["time_in_ms"]]

total_time = pd.DataFrame(execution_times, index=algorithms, columns=["total_execution_time"])
total_time
```

### Plotting DC Algorithm

```python
sns.lineplot(data=troco_DC40, x="value", y="time_in_ms")
plt.xlabel("Valor do Troco")
plt.ylabel("Tempo de Execução (ms)")
plt.savefig("DC_time.png")
```

```python
sns.lineplot(data=troco_DC40, x="value", y="time_in_ms")
plt.yscale("log")
plt.xlabel("Valor do Troco")
plt.ylabel("Tempo de Execução (ms)")
plt.savefig("DC_time_log.png")
```

### Plotting GR Algorithm

```python
sns.lineplot(data=troco_GR_BR200, x="value", y="time_in_ms")
plt.xlabel("Valor do Troco")
plt.ylabel("Tempo de Execução (ms)")
plt.savefig("GR_time.png")
```

### Plotting PD Algorithm

```python
sns.lineplot(data=troco_PD_BR200, x="value", y="time_in_ms")
plt.xlabel("Valor do Troco")
plt.ylabel("Tempo de Execução (ms)")
plt.savefig("PD_time.png")
```

### Plotting the GR Loses DataFrames

```python
troco_GR_GRLoses["Algoritmo"] = 'GR'
troco_PD_GRLoses["Algoritmo"] = 'PD'

GR_loses = pd.concat([troco_GR_GRLoses, troco_PD_GRLoses])
```

```python
sns.lineplot(data=GR_loses, x="value", y="coins_number", hue="Algoritmo")
plt.xlabel("Valor do Troco")
plt.ylabel("Número de Moedas")
plt.savefig("GR_loses_comparation.png")
```
